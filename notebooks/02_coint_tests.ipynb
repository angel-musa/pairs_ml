# 02 Cointegration Tests
Run Engle-Granger test.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import pandas as pd
from core.data_loader import load_price_csv, align_pairs
from core.coint import engle_granger
```

```python
df = load_price_csv("../data/prices.csv", date_col="date")
df_pairs = align_pairs(df, ["AAA", "BBB"])
y = df_pairs["BBB"]
x = df_pairs["AAA"]
```

```python
res = engle_granger(y, x)
print(f"Hedge Ratio: {res.hedge_ratio}")
print(f"ADF Stat: {res.adf_stat}")
print(f"P-Value: {res.adf_pvalue}")
```

```python
res.spread.plot(title="Spread")
```
